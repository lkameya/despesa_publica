```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

plt.style.use('fivethirtyeight')
plt.rcParams['figure.figsize'] = (30, 10)
```

```python
df = pd.read_csv('dados_limpos_ceaps.csv', parse_dates=['DATA'])
```

```python
df.dtypes
```

```python
df.describe()
```

```python
romario = df[df['SENADOR'] == 'ROMÁRIO']
```

```python
romario.head()
```

```python
romario.shape
```

```python
sns.countplot(data=romario, x='TIPO_DESPESA')
plt.title('QUANTIDADE DE DESPESAS LANÇADAS')
plt.xlabel('TIPO DE DESPESA')
plt.ylabel('TOTAL DE DESPESAS')
plt.show()
```

### Como podemos observar no gráfico os maiores gastos foram com Passagens

```python
valores_romario = romario.groupby('TIPO_DESPESA')['VALOR_REEMBOLSADO'].sum()
```

```python
valores_romario
```

```python
valores_romario.plot(kind='line', rot=0, color='green')
```

```python
df['SENADOR'].value_counts().plot(kind='bar')
```

```python
valores_senadores = df.groupby('SENADOR')['VALOR_REEMBOLSADO'].sum().sort_values(ascending=False)
```

```python
valores_senadores
```

```python
df['DATA'].dropna(inplace=True)
```

```python
df['DATA'] = df['DATA'].apply(lambda x: x.strftime('%Y-%m'))
```

```python
gastos_por_mes = df.pivot_table(index='DATA', values='VALOR_REEMBOLSADO', aggfunc=np.sum)
gastos_por_mes.plot(kind='bar')
plt.show()
```

```python
media = df.pivot_table(index='DATA', values='VALOR_REEMBOLSADO', aggfunc=np.mean)
media.columns = ['MEDIA']
mediana = df.pivot_table(index='DATA', values='VALOR_REEMBOLSADO', aggfunc=np.median)
mediana.columns = ['MEDIANA']

media_mediana = pd.concat([media, mediana], axis=1)
media_mediana.head()
```

```python
media.columns
```

```python
media_mediana.plot()
plt.show()
```

```python
dados2017 = df[(df['DATA'] >= '2017-01') & (df['DATA'] <= '2017-12')]
dados2017.tail()
```

```python
mapa_de_calor = dados2017.pivot_table(index='DATA', values='VALOR_REEMBOLSADO', columns='TIPO_DESPESA', aggfunc=np.sum)
mapa_de_calor
```

```python
sns.heatmap(data=mapa_de_calor, annot=True, fmt='g', cmap='Blues')
plt.show()
```
